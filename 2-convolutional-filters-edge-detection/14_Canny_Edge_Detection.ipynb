```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2

%matplotlib inline

!git clone -l -s https://github.com/bttrung/CVND-udacity-facebook-scholarship.git
%cd cloned-repo
!ls
```

```
img_url = 'CVND-udacity-facebook-scholarship/convolutional-filters-edge-detection/images/sunflower.jpg'
image = cv2.imread(img_url)
# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

## TODO: Define lower and upper thresholds for hysteresis
# right now the threshold is so small and low that it will pick up a lot of noise
lower = 120
upper = 240

edges = cv2.Canny(gray, lower, upper)

plt.figure(figsize=(20,10))
plt.imshow(edges, cmap='gray')
```

```
# Try Canny using "wide", "medium" and "tight" thresholds
low = cv2.Canny(gray, 10, 50)
wide = cv2.Canny(gray, 30, 100)
medium = cv2.Canny(gray, 100, 200)
tight = cv2.Canny(gray, 200, 240)
more_tight = cv2.Canny(gray, 230, 250) 
    
# Display the images
f, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(1, 5, figsize=(20,10))

ax0.set_title('low')
ax0.imshow(low, cmap='gray')

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('medium')
ax2.imshow(medium, cmap='gray')

ax3.set_title('tight')
ax3.imshow(tight, cmap='gray')

ax4.set_title('more_tight')
ax4.imshow(more_tight, cmap='gray')
```
