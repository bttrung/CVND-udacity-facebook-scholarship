```
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```
hall = 'CVND-udacity-facebook-scholarship/convolutional-filters-edge-detection/images/city_hall.jpg'
```

```
# Read in an image
image = cv2.imread(hall)

image_copy = np.copy(image)
# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```
# 3x3 array
# Sobel filter
kenel = np.array([[-1,0,1],
                 [-2,0,2],
                 [-1,0,1]])
```

```
kenel
```

```
filtered_image = cv2.filter2D(gray, -1, kenel)
plt.imshow(filtered_image, cmap='gray')
```

## Create binary image

```
retval, binary_image = cv2.threshold(filtered_image, 150, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
```

```
# try to apply another kenel
kenel = np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])
```

```
filtered_image = cv2.filter2D(gray, -1, kenel)
plt.imshow(filtered_image, cmap='gray')
```

```
kenel = np.array([[-1,-5,-1],
                    [0,0,0],
                   [1,5,1]])
```

```
filtered_image = cv2.filter2D(gray, -1, kenel)
plt.imshow(filtered_image, cmap='gray')
```

```
kenel = np.array([[-10,0,10],
                 [-50,0,50],
                 [-10,0,10]])

filtered_image = cv2.filter2D(gray, -1, kenel)
plt.imshow(filtered_image, cmap='gray')
```

```
kenel = np.array([[-1,0,1],
                 [-1,0,1],
                 [-1,0,1]])

filtered_image = cv2.filter2D(gray, -1, kenel)
plt.imshow(filtered_image, cmap='gray')
```
