```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2

%matplotlib inline
```

```
lane = 'CVND-udacity-facebook-scholarship/convolutional-filters-edge-detection/images/curved_lane.jpg'
image = mpimg.imread(lane)
plt.imshow(image)
```

```
# Convert to grayscale for filtering
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```
# Create a custom kernel
# 3x3 array for edge detection
sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

sobel_x = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])
```

```
# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image_by_y = cv2.filter2D(gray, -1, sobel_y)
filtered_image_by_x = cv2.filter2D(gray, -1, sobel_x)
```

```
plt.imshow(filtered_image_by_y, cmap='gray')
```

```
plt.imshow(filtered_image_by_x, cmap='gray')
```

```
# let try to apply both filters
# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image_by_y = cv2.filter2D(gray, -1, sobel_y)
filtered_image_by_x = cv2.filter2D(filtered_image_by_y, -1, sobel_x)
plt.imshow(filtered_image_by_x, cmap='gray')
```
